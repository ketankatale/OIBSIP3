```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import calendar
import datetime as dt
%matplotlib inline
```

```python
ui1 = pd.read_csv('/content/Unemployment_Rate_upto_11_2020.csv')
ui1.head(10)
```

```python
ui1.columns = ['States','Date','Frequency','Unemployment Rate','Employed','Labour Participation Rate','Region','longitude','latitude']
ui1
```

```python
ui1['Date'] = pd.to_datetime(ui1['Date'])
```

```python
ui1['month_int'] = ui1['Date'].dt.month
```

```python
ui1['month'] = ui1['month_int'].apply(lambda x: calendar.month_abbr[x])
```

```python
ui1['Region'].hist()
plt.title("Region unemployment Analysis by histogram")
```

```python
region_unemployment = ui1.groupby('Region')['Unemployment Rate'].mean().reset_index()
figure = px.pie(region_unemployment, names='Region', values='Unemployment Rate',
                title='Average Unemployment Rate by Region in India')
figure.show()
```

```python
ur= px.bar(ui1, x='States', y='Unemployment Rate', color='States', title='Unemployment Rate (States wise) by Bar Graph', template='plotly')
ur.update_xaxes(categoryorder='total descending')
ur.show()
```

```python
dig = px.scatter_matrix(ui1, template='plotly',
                        dimensions=['Unemployment Rate', 'Employed', 'Labour Participation Rate'],
                        color='Region')
dig.show()
```

```python
ur= px.box(ui1, x='Region', y='Unemployment Rate', color='Region', title='Unemployment Rate (Region wise) by Box plot', template='plotly')
ur.update_xaxes(categoryorder='total descending')
ur.show()
```

```python
ur= px.violin(ui1, x='States', y='Employed', color='States', title='employment Rate (state wise)', template='seaborn')
ur.update_xaxes(categoryorder='total descending')
ur.show()
```

```python
plt.style.use('seaborn-talk')
plt.figure(figsize=(10, 7))
ax = plt.subplot()
ax.tick_params(axis='both', which='both', length=12, labelsize=12, labelcolor='black')
sns.heatmap(ui1.corr(), annot=True, linewidths=4, cmap='RdYlBu')
plt.title('Correlation between dataset by Heatmap', fontsize=16)
plt.show()
```

```python
unemp= ui1[['States', 'Region', 'Unemployment Rate','month']]
dig = px.sunburst(unemp,
                     path=['Region', 'States', 'month'],
                     values='Unemployment Rate',
                     width=1200,
                     height=1000,
                     color='Unemployment Rate',
                     color_continuous_scale='RdYlGn',
                     title='Unemployment Rate in India')
dig.show()
```

```python
# Assuming ui1 is your DataFrame
plt.figure(figsize=(8, 6))
plt.scatter(x=ui1['Unemployment Rate'], y=ui1['Employed'], marker='o', color='blue', label='Data Points')
plt.title('Unemployment vs Employment')
plt.xlabel('Unemployment percentage')
plt.ylabel('Employment percentage')
plt.legend()  # If you want to display the legend for the data points
plt.grid(True)  # If you want to display a grid on the plot
plt.show()
```
